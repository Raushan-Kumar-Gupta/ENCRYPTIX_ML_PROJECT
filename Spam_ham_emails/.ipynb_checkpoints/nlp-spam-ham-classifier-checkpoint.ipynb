```python
df=pd .read_csv("/kaggle/input/sms-spam-collection-dataset/spam.csv",encoding='latin-1')
df.head()
```

```python
data=df.drop(df.columns[2:],axis=1)
data=data.rename(columns={'v1':'target','v2':'message'})
data
```

```python
import nltk
import re
```

## One Hot Encoding

##### we apply get_dummies function to the categorical variable 'target' to convert it into one hot encoded columns ham/spam 

```python
dummies=pd.get_dummies(data['target'])
dummies
```

```python
data=pd.concat([data,dummies],axis=1)
data.head()
```

```python
data['ham'] = data['ham'].astype(int)
data['spam'] = data['spam'].astype(int)
data
```

```python
def message(text):
        text=re.sub('[^a-zA-Z0-9]'," ",text)
        text=text.lower()
        return text
    
```

```python
data['message_updated']=data['message'].apply(message)
data.head()
```

```python
data['message_length']=data['message'].apply(len)
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot (data['target'])
plt.title("Dataset Distribution")
plt.show()
```

#### The Dataset is imbalanced as 'ham' category is distributed much more than 'spam' category. We can apply various techniques to handle imbalanced dataset - but here we are mainly focusing on NLP techniques

```python
data.hist(column='message_length',by='target',bins=60,figsize=(12,4));
```

#### As we can see that message length of spam messages is more than message length of ham messages

## Stop Words

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
```

```python

def remove_stop_words(text):
    text = ' '.join(word for word in text.split(' ') if word not in stop_words)
    return text
```

```python
data['message_updated']=data['message_updated'].apply(remove_stop_words)
```

```python
data.message_updated[0]
```

```python
data
```

## Stemming

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
```

```python
def stemming(text):
    text=stemmer.stem(text)
    return text
```

```python
data['message_updated']=data['message_updated'].apply(stemming)
```

```python
data
```

```python

y=dummies.iloc[:,1].values
y
```

```python
x=data.loc[:,'message_updated']
x
```

```python
list_=x.tolist()
list_
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
x=cv.fit_transform(list_).toarray()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=30)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()

# Train the model
nb.fit(x_train, y_train)
```

```python
y_pred=nb.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test, y_pred))

cm=metrics.confusion_matrix(y_test, y_pred)
```

```python
cm_df = pd.DataFrame(cm, index=['Ham', 'Spam'], columns=['Ham', 'Spam'])

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

#### As we can see from the confusion matrix that there are 15 instances when actually message was a ham but was classified as spam which means we can further optimize the model to reduce such instances because this is an important scenario where a ham message should not be classified as spam as there could be chances of missing important messages.
